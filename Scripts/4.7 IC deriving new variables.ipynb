```python
# Create 'price_label' function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Add 'price_label' to dataframe

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check value of product with maximum price in dataframe

df['prices'].max()
```

```python
# Use loc() function to achieve same result

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## Q4. Create a new 'busiest_period_of_day' column

```python
# Identify busiest hours of the day

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Use a for-loop to create a new 'busiest_period_of_day' column

result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if (value >= 9) and (value <= 17):
    result.append("Most orders")
  elif (value >= 20) | (value < 9):
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
# Add new column to dataframe

df_ords_prods_merged['busiest_period_of_day'] = result
```
