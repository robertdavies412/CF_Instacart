```python
# Create a list of randomized values to prepare to split the dataframe for sampling

np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# Create a big dataframe with a sample of 70% of the main dataframe and a small dataframe with 30%

big = df[dev]
small = df[~dev]
```

```python
# Check output

len(df)
```

```python
len(big) + len(small)
```

```python
# Reduce sample to only columns needed from small sample

df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_dow_sample.png'))
```
